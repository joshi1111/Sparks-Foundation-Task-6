Encoding Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
dataset['Species']=le.fit_transform(dataset['Species'])
dataset.head()
```

Correlation Matrix

```
dataset.drop(['Id'], axis=1,inplace=True)
corrMatrix = dataset.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

Splitting Train and Test Dataset

```
x=dataset.iloc[:,0:4].values
y=dataset.iloc[:,-1].values

from sklearn.model_selection import  train_test_split
x_train, x_test, y_train, y_test=  train_test_split( x, y , test_size=0.2, random_state=23)
print(x_train)
print(x_test)
```

Training the dataset

```
from sklearn.tree import DecisionTreeClassifier
classifier= DecisionTreeClassifier(random_state=23)
classifier.fit(x_train,y_train)
```

```
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

fn= np.array(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm'])
cln= np.array(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
dot_data = StringIO()
export_graphviz(classifier, out_file=dot_data, feature_names=fn, class_names= cln,
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

Accuracy

```
from sklearn.metrics  import accuracy_score, confusion_matrix

y_pred_train= classifier.predict(x_train)

cm2= confusion_matrix(y_train,y_pred_train)
print('Accuracy Score of Train Set:')
print(accuracy_score(y_train,y_pred_train))
print(cm2)
```

```
from sklearn.metrics import classification_report
y_pred= classifier.predict(x_test)

cm1= confusion_matrix(y_test,y_pred)

print('Accuracy Score of test set:') 
print(accuracy_score(y_test,y_pred))
print(cm1)
print(classification_report(y_test, y_pred, target_names=cln))
```

```
result= pd.DataFrame({'Actual':le.inverse_transform(y_test), 'Predicted': le.inverse_transform(y_pred)})
status=[]
for ind in result.index: 
     if (result['Actual'][ind] == result['Predicted'][ind]):
       status.append('Correct')
     else:
       status.append('Incorrect')


result['Status']= status

print(result)
```
